## PREGUNTA 3 (2pt)
Crear una **VISTA** (view) llamada "Valor" conteniendo el valor total del pedido (1 tupla) y desplegarla, la vista contiene los siguientes atributos:
idCliente, el número identificador del cliente
Valor, consistente en el valor total del pedido

Creamos la tabla P44 para solo seleccionar aquellos pedidos del cliente 44, que es el id que nos interesa

```python
%%sql
DROP TABLE IF EXISTS P44;
CREATE TABLE P44 AS
SELECT cliente, fecha, hora, unnest(string_to_array(plato, ' ')) AS plato_id
FROM pedido
WHERE cliente = 44;

SELECT * FROM P44;
```

```python
%%sql
DROP VIEW IF EXISTS Vista44;
CREATE VIEW Vista44 AS
SELECT
    cliente AS idCliente,
    SUM(pl.precio) AS Valor
FROM
    P44
JOIN
    plato pl ON P44.plato_id::integer = pl.id
GROUP BY
    cliente;
SELECT * FROM Vista44;
```

## Pregunta 5 (2pt)
Crear un **STORED PROCEDURE** llamado "ListaPedidos" que genere una lista de todos los pedidos de un cliente ingresado como parámetro y desplegarla ordenado por fecha (no hora)



```python
%%sql

DROP PROCEDURE IF EXISTS ListaPedidos;
```

```python
%%sql

CREATE OR REPLACE FUNCTION ListaPedidos(cliente_id INT)
RETURNS TABLE(id INT, cliente INT, plato VARCHAR, fecha DATE, hora TIME) AS $$
BEGIN
    RETURN QUERY
    SELECT pedido.id, pedido.cliente, pedido.plato, pedido.fecha, pedido.hora
    FROM pedido
    WHERE pedido.cliente = cliente_id
    ORDER BY pedido.fecha;
END;
$$ LANGUAGE plpgsql;
```

Para probarlo ejecute SELECT * FROM ListaPedidos (numero-de-grupo);

```python
%%sql

SELECT * FROM ListaPedidos(44);
```
